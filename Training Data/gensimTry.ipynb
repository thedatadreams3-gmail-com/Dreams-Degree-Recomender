# Gensim Train word2vec

```python
import gensim
import os

model = gensim.models.Word2Vec(documents, size=150, window=10, min_count=2, workers=4)
```

```python
model.most_similar('programming')
```

```python
model.wv['coding']
```

# DOC2VEC

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
docs = [TaggedDocument(doc, [i]) for i, doc in enumerate(documents)]
```

```python
docs
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

modelDoc= gensim.models.Doc2Vec(docs, vector_size=150, window=15, min_count=2, workers=4, epochs=40)
```

# Test Using Text

```python
words='i like mathemathics, programming, coding logic'.split()

new_vec=modelDoc.infer_vector(words)
new_vec
```

```python
#similar_doc = modelDoc.docvecs.most_similar([new_vec], topn=5)
similar_doc = modelDoc.docvecs.most_similar([new_vec])
```

```python
pred = []
for row,index in similar_doc:
    pred.append(
        {
            'Study Program':df['Study Program'][row] ,
            'Courses': df['Courses'][row],
            'Similarity':  str(index)
        }
    )

pred=pd.DataFrame(pred)
```
